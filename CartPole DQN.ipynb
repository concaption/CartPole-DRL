### Import dependences

```python
# !pip install stable-baselines3[extra]
import os
import gym
from stable_baselines3 import DQN
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

### Load Environment

```python
# MAKE YOUR DIRECTORIES FIRST
log_path = os.path.join('Training','Logs')
```

```python
environment_name = 'CartPole-v0'
env = gym.make(environment_name)
env = DummyVecEnv([lambda: env])
model = DQN(policy='MlpPolicy', env=env, verbose=1, tensorboard_log=log_path)
```

### Train Agent

```python
model.learn(total_timesteps=200000)
```

### Save Model

```python
saved_path = os.path.join('Training','Saved Models', 'DQN_CartPole.zip')
model.save(saved_path)
```

### Load Model

```python
del model
model = DQN.load(saved_path,env= env)
```

### Evaluation

```python
evaluate_policy(model, env,n_eval_episodes=10,render=True)
env.close()
```

#### Test Model

```python
episodes = 5
for episode in range(1, episodes+1):
    obs = env.reset()
    done = False
    score = 0
    for step in range(200):
        env.render()
        action, _states = model.predict(obs) # now using the model
        obs, rewards, done, info = env.step(action)
        score += rewards
        if done:
            break
    print(f'Episode {episode} finished after {step} timesteps')
    
env.close()
```
